```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
mse_list = []
for layer in [1,4,5,7]:
    filename1 = "cifar10_vgg_trail1/weights/"+str(layer)+".csv"
    filename2 = "cifar10_vgg_trail2/weights/"+str(layer)+".csv"
    file1 = pd.read_csv(filename1,header=None)
    file2 = pd.read_csv(filename2,header=None)
    wei1 = file1.tail(1).values[0]
    wei2 = file2.tail(1).values[0]
    scale_wei1 = wei1 - min(wei1) / max(wei1) - min(wei1)
    scale_wei2 = wei2 - min(wei2) / max(wei2) - min(wei2)
    mse = ((scale_wei1 - scale_wei2)**2).mean()
    mse_list.append(mse)
```

```python
plt.figure()
plt.plot([1,4,5,7],mse_list,'--')
plt.xlabel("Layer")
plt.ylabel("MSE")
plt.title("MSE of final parameter values between 2 trails")
```
